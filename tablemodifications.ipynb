<h3>Updating values </h3>
<p> Updating a player's team, here we change the the team that Kevin Durant is on in 2019 to the Celtics </p>

```python
%%sql
UPDATE Players SET TEAM_ID=1610612738 WHERE PLAYER_NAME = 'Kevin Durant' and SEASON=2019;
SELECT * FROM Players WHERE PLAYER_NAME = "Kevin Durant" and SEASON=2019;
```

<p> Notice that in the following query since we have changed Kevin Durant to play for Team ID 1610612738 that update will be reflected in the TEAM_ID column due to our update trigger. As well our Update_Player_Team_City trigger will change the TEAM_CITY column to Boston using a subquery </p>

```python
%%sql 

SELECT * FROM Games_Details JOIN Games on Games_Details.GAME_ID = Games.GAME_ID WHERE PLAYER_NAME = "Kevin Durant"LIMIT 10;
```
