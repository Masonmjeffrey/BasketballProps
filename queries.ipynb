<h3> Team Roster </h3>
We join the player table with the teams table, then select Boston, this is the Team Roster for the Boston Celtics

```python
%%sql

    SELECT p.PLAYER_NAME, t.CITY, t.NICKNAME, p.season
    FROM Players p
    LEFT JOIN Teams t on p.TEAM_ID = t.TEAM_ID
    WHERE CITY = "BOSTON"
    ORDER BY season DESC
    LIMIT 20;
```

<h3> Career Averages </h3>
This contains the career averages for each player across each statistical category. We join games details with games and grab the average of each statisticial category

```python
%%sql

    SELECT * FROM
        (SELECT gd.PLAYER_NAME, AVG(gd.PTS) as AVERAGE_POINTS, AVG(gd.AST) as AVERAGE_ASSISTS, AVG(gd.REB) as AVERAGE_REBOUNDS, AVG(gd.FG_PCT)*100 as AVERAGE_FG_PCT, AVG(gd.FG3_PCT)*100 as AVERAGE_3PT_PCT, AVG(FT_PCT)*100 as AVERAGE_FT_PCT
        FROM Games_Details gd
        GROUP BY gd.PLAYER_NAME) pp
    WHERE AVERAGE_POINTS IS NOT NULL
        AND AVERAGE_ASSISTS IS NOT NULL
        AND AVERAGE_REBOUNDS IS NOT NULL
        AND AVERAGE_FG_PCT IS NOT NULL
        AND AVERAGE_3PT_PCT IS NOT NULL
        AND AVERAGE_FT_PCT IS NOT NULL
    ORDER BY pp.AVERAGE_POINTS DESC
    LIMIT 20;
```

<h3> Players who averaged 10,5,5 </h3>
<p> This table contains the career averages of players who averaged 10 points, 5 assists, and 5 rebounds over their NBA careers </p>

```python
%%sql

    SELECT * FROM
        (SELECT gd.PLAYER_NAME, AVG(gd.PTS) as AVERAGE_POINTS, AVG(gd.AST) as AVERAGE_ASSISTS, AVG(gd.REB) as AVERAGE_REBOUNDS
        FROM Games_Details gd
        GROUP BY gd.PLAYER_NAME) pp
    WHERE pp.AVERAGE_POINTS > 10.0 AND pp.AVERAGE_ASSISTS > 5.0 AND pp.AVERAGE_REBOUNDS > 5.0
    ORDER BY pp.AVERAGE_POINTS DESC;
```

<h3> Best Individual Seasons </h3>
<p> This table joins Games on Games Details and groups the information by PLAYER_NAME and SEASONS, thus it represents the best individual seasons</p>

```python
%%sql

SELECT * FROM
        (SELECT gd.PLAYER_NAME, AVG(gd.PTS) as AVERAGE_POINTS, AVG(gd.AST) as AVERAGE_ASSISTS, AVG(gd.REB) as AVERAGE_REBOUNDS, g.SEASON
        FROM Games_Details gd
        LEFT JOIN Games g on gd.GAME_ID = g.GAME_ID
        GROUP BY gd.PLAYER_NAME, g.SEASON) pp
    WHERE pp.AVERAGE_POINTS > 10.0 AND pp.AVERAGE_ASSISTS > 5.0
    ORDER BY pp.AVERAGE_POINTS DESC
    LIMIT 25;
```

<h3> Season Averages for Steph Curry </h3>
<p> This table joins Games on Games_Details, and filters to get Steph Curry's records. It then groups by season to compute the averages across each season. </p>

```python
%%sql
    
    SELECT gd.PLAYER_NAME, AVG(gd.PTS) as AVERAGE_POINTS, AVG(gd.AST) as AVERAGE_ASSISTS, AVG(gd.REB) as AVERAGE_REBOUNDS, AVG(gd.FG_PCT)*100 as AVERAGE_FIELD_GOAL_PCT, AVG(gd.FG3_PCT)*100 as AVERAGE_3POINT_PCT, AVG(gd.FT_PCT)*100 as AVERAGE_FT_PCT, g.SEASON
    FROM Games_Details gd
    LEFT JOIN Games g on gd.GAME_ID = g.GAME_ID
    WHERE PLAYER_NAME = "Stephen Curry"
    GROUP BY g.SEASON
    ORDER BY AVERAGE_POINTS DESC
    LIMIT 10;
```

<h3> Win Percentage of a Specified Team against another Specified Team </h3>
<p> This is the second of the main functionalities calculating the win percentage of one specified team against another specified team. Here we calculate the win percentage of Chicago against Philly over the last 10 games that they played. </p>
<p> First we create two CTEs one for chicago GAME_IDs and one for philly Game_IDs </p>
<p> Next we create a CTE consisting of the last 10 games played between chicago and philly </p>
<p> We then create a CTE that contains a count of the number of wins each team had at home and away </p>
<p> Lastly we calculate the percentage of one team's wins over the last ten games as the sum of their wins at home plus their wins away divided by 10 </p>

```python
%%sql

    with chicago_games as (SELECT GAME_ID FROM Games
    WHERE TEAM_ID_home = (SELECT TEAM_ID 
    FROM Teams t
    WHERE NICKNAME = "Bulls") OR 
    TEAM_ID_away = (SELECT TEAM_ID 
    FROM Teams t
    WHERE NICKNAME = "Bulls")),
    philly_games as (SELECT GAME_ID FROM Games
    WHERE TEAM_ID_home = (SELECT TEAM_ID 
    FROM Teams t
    WHERE NICKNAME = "76ers") OR 
    TEAM_ID_home = (SELECT TEAM_ID 
    FROM Teams t
    WHERE NICKNAME = "76ers"))


    SELECT * 
    FROM Games g 
    WHERE g.GAME_ID in (SELECT * FROM philly_games) AND g.GAME_ID in (SELECT * FROM chicago_games)
    ORDER BY g.GAME_DATE_EST DESC
    LIMIT 10;
```

```python
%%sql

    with chicago_games as (SELECT GAME_ID FROM Games
    WHERE TEAM_ID_home = (SELECT TEAM_ID 
    FROM Teams t
    WHERE NICKNAME = "Bulls") OR 
    TEAM_ID_away = (SELECT TEAM_ID 
    FROM Teams t
    WHERE NICKNAME = "Bulls")),
    philly_games as (SELECT GAME_ID FROM Games
    WHERE TEAM_ID_home = (SELECT TEAM_ID 
    FROM Teams t
    WHERE NICKNAME = "76ers") OR 
    TEAM_ID_home = (SELECT TEAM_ID 
    FROM Teams t
    WHERE NICKNAME = "76ers")),
    last_ten_philly_chicago as (
        SELECT * FROM Games g 
        WHERE g.GAME_ID in (SELECT * FROM philly_games) AND g.GAME_ID in (SELECT * FROM chicago_games)
        ORDER BY g.GAME_DATE_EST DESC
        LIMIT 10),
    philly_home_wins as (
        SELECT COUNT(HOME_TEAM_WINS) as wins
        FROM last_ten_philly_chicago
        WHERE last_ten_philly_chicago.HOME_TEAM_WINS = 1 and last_ten_philly_chicago.TEAM_ID_home = (SELECT TEAM_ID FROM Teams WHERE NICKNAME = "76ers")),
    philly_away_wins as (
        SELECT COUNT(HOME_TEAM_WINS) as wins
        FROM last_ten_philly_chicago
        WHERE last_ten_philly_chicago.HOME_TEAM_WINS = 0 and last_ten_philly_chicago.TEAM_ID_away = (SELECT TEAM_ID FROM Teams WHERE NICKNAME = "76ers")),
    chicago_home_wins as (
        SELECT COUNT(HOME_TEAM_WINS) as wins
        FROM last_ten_philly_chicago
        WHERE last_ten_philly_chicago.HOME_TEAM_WINS = 1 and last_ten_philly_chicago.TEAM_ID_home = (SELECT TEAM_ID FROM Teams WHERE NICKNAME = "Bulls")),
    chicago_away_wins as (
        SELECT COUNT(HOME_TEAM_WINS) as wins
        FROM last_ten_philly_chicago
        WHERE last_ten_philly_chicago.HOME_TEAM_WINS = 0 and last_ten_philly_chicago.TEAM_ID_away = (SELECT TEAM_ID FROM Teams WHERE NICKNAME = "Bulls")
    )
    
    SELECT (sum(wins)/10)*100 "Chicago Win Percentage"
    FROM ( SELECT * FROM chicago_away_wins UNION ALL SELECT * FROM chicago_home_wins ) chicago_wins;
```
