<h3> Tables </h3>

First we join the player table with the teams table, this represents the roster for a given season for each team

```python
# Table player with their team for each season
q = """
    SELECT p.PLAYER_NAME, t.CITY, t.NICKNAME, p.season
    FROM players p
    LEFT JOIN teams t on p.TEAM_ID = t.TEAM_ID
    LIMIT 20;"""
playersdf = pysqldf(q)
playersdf
```

This contains the career averages for each player across each statistical category.

```python
# Table with player career averages
q = """
    SELECT * FROM
        (SELECT gd.PLAYER_NAME, AVG(gd.PTS) as AVERAGE_POINTS, AVG(gd.AST) as AVERAGE_ASSISTS, AVG(gd.REB) as AVERAGE_REBOUNDS, AVG(gd.FG_PCT)*100 as AVERAGE_FG_PCT, AVG(gd.FG3_PCT)*100 as AVERAGE_3PT_PCT, AVG(FT_PCT)*100 as AVERAGE_FT_PCT
        FROM games_details gd
        GROUP BY gd.PLAYER_NAME) pp
    WHERE AVERAGE_POINTS IS NOT NULL
        AND AVERAGE_ASSISTS IS NOT NULL
        AND AVERAGE_REBOUNDS IS NOT NULL
        AND AVERAGE_FG_PCT IS NOT NULL
        AND AVERAGE_3PT_PCT IS NOT NULL
        AND AVERAGE_FT_PCT IS NOT NULL
    ORDER BY pp.AVERAGE_POINTS DESC
    LIMIT 20;
    """
player_game_pt_asst = pysqldf(q)
player_game_pt_asst
```

Players who averaged more than 10 points and 5 assists

```python
# This should be the players who averaged the most points over their career
# among players in the dataset who averaged more than 10 points and 5 assists
# This should contain no duplicate players as I only grouped by player and it
# represents career averages
q = """
    SELECT * FROM
        (SELECT gd.PLAYER_NAME, AVG(gd.PTS) as AVERAGE_POINTS, AVG(gd.AST) as AVERAGE_ASSISTS, AVG(gd.REB) as AVERAGE_REBOUNDS
        FROM games_details gd
        GROUP BY gd.PLAYER_NAME) pp
    WHERE pp.AVERAGE_POINTS > 10.0 AND pp.AVERAGE_ASSISTS > 5.0
    ORDER BY pp.AVERAGE_POINTS DESC;
    """
player_game_pt_asst = pysqldf(q)
player_game_pt_asst
```

This next table represents the top player seasons ordered by points, this table should have duplicates as one player can have multiple great seasons.

```python
# This should be the top 25 player seasons ordered by average points who averaged
# more than 10 points and 5 assists per game in a given season.
# In this query I expect duplicate players as there are players who have met this
# criteria in more than one season. I do expect a different year if the player 
# is the same though.
# 
q = """
    SELECT * FROM
        (SELECT gd.PLAYER_NAME, AVG(gd.PTS) as AVERAGE_POINTS, AVG(gd.AST) as AVERAGE_ASSISTS, AVG(gd.REB) as AVERAGE_REBOUNDS, g.SEASON
        FROM games_details gd
        LEFT JOIN games g on gd.GAME_ID = g.GAME_ID
        GROUP BY gd.PLAYER_NAME, g.SEASON) pp
    WHERE pp.AVERAGE_POINTS > 10.0 AND pp.AVERAGE_ASSISTS > 5.0
    ORDER BY pp.AVERAGE_POINTS DESC
    LIMIT 25;
    """
player_game_season = pysqldf(q)
player_game_season
```

This next table shows the season averages for a player across each individual season in the significant statistical cateogry.

```python
# These are the season averages for an indiviudal player, this grouping is done by
# season to show how a player performed across varoius different seasons
# I just chose Stephen Curry but ideallyl the user will select 
q = """
    SELECT gd.PLAYER_NAME, AVG(gd.PTS) as AVERAGE_POINTS, AVG(gd.AST) as AVERAGE_ASSISTS, AVG(gd.REB) as AVERAGE_REBOUNDS, AVG(gd.FG_PCT)*100 as AVERAGE_FIELD_GOAL_PCT, AVG(gd.FG3_PCT)*100 as AVERAGE_3POINT_PCT, AVG(gd.FT_PCT)*100 as AVERAGE_FT_PCT, g.SEASON
    FROM games_details gd
    LEFT JOIN games g on gd.GAME_ID = g.GAME_ID
    WHERE PLAYER_NAME = "Stephen Curry"
    GROUP BY g.SEASON
    ORDER BY AVERAGE_POINTS DESC
    LIMIT 10;
    """
individual_game_season = pysqldf(q)
individual_game_season
```

<h3> Player Props Main Query </h3>
Now onto the big query finding an individual player's statistical performances against a specific team over the last 10 times that they've played them. First we are going to find all the games where the brooklyn nets played in.

```python
# This is a CTE of all games that the Brooklyn Nets played
q = """
    with nets_games as (SELECT GAME_ID FROM games
    WHERE HOME_TEAM_ID = (SELECT TEAM_ID 
    FROM teams t
    WHERE NICKNAME = "Nets") OR 
    VISITOR_TEAM_ID = (SELECT TEAM_ID 
    FROM teams t
    WHERE NICKNAME = "Nets"))
    SELECT * FROM nets_games LIMIT 10;

    """
team_info = pysqldf(q)
team_info
```

Next we want to find all of the games that Jaylen Brown played in, this is relatively simple but we will join with games here because we want the Game Date.

```python
# Now I want to find all games that Jaylen Brown played in,
# should be easy enough
q = """
    SELECT g.GAME_DATE_EST, gd.PLAYER_NAME, gd.PTS, gd.REB, gd.AST, gd.FG_PCT, gd.FG3_PCT, gd.FT_PCT
    FROM games_details gd LEFT JOIN games g ON gd.GAME_ID = g.GAME_ID
    WHERE gd.PLAYER_NAME = "Jaylen Brown"
    ORDER BY GAME_DATE_EST DESC
    LIMIT 10;
    """
individual_games_stats = pysqldf(q)
individual_games_stats
```

Now we can combine these two queries and get our target. This will contain the statistical performances of Jaylen Brown over the last 10 games that he played against the Brooklyn Nets.

```python
# Now we need to do matchups for one player
# for this one we want to find how Jaylen Brown has performed
# against the Brookyln Nets the last 10 times that he has played them
# we will simply used the CTE that we defined earlier and then only stats for Jaylen Brown in those games
q = """
    with nets_games as (SELECT GAME_ID FROM games
    WHERE HOME_TEAM_ID = (SELECT TEAM_ID 
    FROM teams t
    WHERE NICKNAME = "Nets") OR 
    VISITOR_TEAM_ID = (SELECT TEAM_ID 
    FROM teams t
    WHERE NICKNAME = "Nets"))

    SELECT g.GAME_DATE_EST, gd.PLAYER_NAME, gd.PTS, gd.REB, gd.AST, gd.FG_PCT, gd.FG3_PCT, gd.FT_PCT
    FROM games_details gd
    LEFT JOIN games g ON gd.GAME_ID = g.GAME_ID
    WHERE gd.PLAYER_NAME = "Jaylen Brown" AND gd.GAME_ID IN nets_games
    ORDER BY GAME_DATE_EST DESC
    LIMIT 10;
    """
player_matchups = pysqldf(q)
player_matchups
```
